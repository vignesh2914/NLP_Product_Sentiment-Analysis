```python
import re
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # Emoticons
                           u"\U0001F300-\U0001F5FF"  # Symbols & Pictographs
                           u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
                           u"\U0001F1E0-\U0001F1FF"  # Flags (iOS)
                           u"\U00002500-\U00002BEF"  # Chinese characters
                           u"\U00002702-\U000027B0"  # Miscellaneous symbols
                           u"\U000024C2-\U0001F251"  # Enclosed characters
                           u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                           "]+", flags=re.UNICODE)
df['review_text'] = df['review_text'].str.replace(emoji_pattern, '', regex=True)
```

```python
df['review_text'] = df['review_text'].str.replace(r'[^a-zA-Z0-9\s]', ' ', regex=True)
```

```python
print("Original shape:", df.shape)
print("Sample data before transformations:")
print(df['review_text'].head(10))
```

```python
import nltk
from nltk.corpus import stopwords


nltk.download('stopwords')
df['review_text'] = df['review_text'].apply(
    lambda x: " ".join([word for word in str(x).split() if word not in stopwords.words('english')]) if isinstance(x, str) else x
)

print("Shape after stopword removal:", df.shape)
print("Sample data after stopword removal:")
print(df['review_text'].head(10))
```

```python
# Remove special characters
df['review_text'] = df['review_text'].apply(
    lambda x: " ".join([word for word in str(x).translate(str.maketrans('', '', '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~')).split() if word not in stopwords.words('english')]) if isinstance(x, str) else x
)

print("Shape after special character removal:", df.shape)
print("Sample data after special character removal:")
print(df['review_text'].head(10))
```

```python
# Replace multiple spaces with a single space
df['review_text'] = df['review_text'].str.replace(r'\s+', ' ', regex=True)

# Final clean-up
df['review_text'] = df['review_text'].str.strip()

print("Final shape:", df.shape)
print("Final sample data:")
print(df['review_text'].head(10))
```

```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')


df['review_text'].fillna('', inplace=True)

# Now you can continue with lemmatization and other preprocessing
lemmatizer = WordNetLemmatizer()
corpus = []

for i in range(len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['review_text'][i])
    review = review.lower()
    review = review.split()

    # Lemmatize each word in the review
    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
nltk.download('punkt')
```

```python
words= []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
def avg_word2vec(doc):
    vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    else:
        # Return a zero vector with the same number of dimensions as your Word2Vec model (e.g., 100 dimensions)
        return np.zeros(model.vector_size)
```
